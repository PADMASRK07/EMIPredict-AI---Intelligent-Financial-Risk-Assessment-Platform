```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report


emi_df = pd.read_csv('D:\Emi Predict\Data Sets\emi_prediction_cleanafterFE.csv')
print(emi_df.shape)
features = [
    "monthly_salary","school_fees","college_fees","travel_expenses",
    "groceries_utilities","other_monthly_expenses",
    "current_emi_amount","credit_score","requested_amount","bank_balance","emergency_fund"
]

X = emi_df[features]
y_clf = emi_df['emi_eligibility']

X_train, X_test, y_train, y_test = train_test_split(
    X, y_clf, test_size=0.2, random_state=42
)
rf_model = RandomForestClassifier(
    n_estimators=100,
    max_depth=10,
    n_jobs=-1,
    random_state=42
)

rf_model.fit(X_train, y_train)

y_pred_rf = rf_model.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred_rf)
precision = precision_score(y_test, y_pred_rf,average='weighted')
recall = recall_score(y_test, y_pred_rf,average='weighted')
f1 = f1_score(y_test, y_pred_rf,average='weighted')

print("📊 Classification Metrics")
print(f"Accuracy  : {accuracy:.4f}")
print(f"Precision : {precision:.4f}")
print(f"Recall    : {recall:.4f}")
print(f"F1 Score  : {f1:.4f}")

# Confusion Matrix
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred_rf))

# Full Report
print("\nClassification Report:")
print(classification_report(y_test, y_pred_rf))
```

```python
with mlflow.start_run(run_name='RF Classification'):
    mlflow.log_metric('accuracy', accuracy)
    mlflow.log_metric('precision', precision)
    mlflow.log_metric('recall', recall)

    mlflow.sklearn.log_model(rf_model,"Random Forest Regressor")
```
